Copy the users and roles DataFrames from the examples above.

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

What is the result of using a right join on the DataFrames?

```python
pd.merge(users, roles, how='right', left_on='role_id', right_on='id')
```

What is the result of using an outer join on the DataFrames?

```python
#indicator argument shows where the data is coming from in merge
pd.merge(users, roles, how='outer', left_on='role_id', right_on='id', indicator=True)
```

What is the total price for each order?

```python
# Convert the 'item_price' column to a numeric data type
chipotle['item_price'] = chipotle['item_price'].str.replace('$', '').astype(float)

# Calculate the total price for each item in the order
chipotle['total_price'] = chipotle['quantity'] * chipotle['item_price']

# Group the data by 'order_id' and sum the total price for each order
order_totals = chipotle.groupby('order_id')['total_price'].sum()

order_totals
```

What are the most popular 3 items?

```python
item_popularity = chipotle.groupby('item_name')['quantity'].sum()
item_popularity
```

```python
item_popularity.sort_values(ascending=False).head(3)
```

Which item has produced the most revenue?

```python
chipotle.groupby('item_name').item_price.sum().nlargest(1)
```

Write the code necessary to create a cross tabulation of the number of titles by department. (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)

```python
from env import host, username, password

url = f'mysql+pymysql://{env.username}:{env.password}@{env.host}/employees'
```

```python
pd.read_sql('show tables', url)
```

```python
query = '''
    select *
    from dept_emp
        join departments
            using (dept_no)
'''
```

```python
df_depts = pd.read_sql(query, url)
df_depts.head()
```

```python
df_emps.head()
```

```python
df_all = pd.merge(df_emps, df_depts, how='inner', on='emp_no')
df_all.head()
```

```python
pd.crosstab(df_all.title, df_all.dept_name)
```
