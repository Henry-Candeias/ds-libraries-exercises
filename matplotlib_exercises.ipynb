Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

y = âˆšx 

y = x^3 

y = 2^x 

y = 1 / (x+1)

You can use functions from the math module to help implement some of the equations above.

```python
import matplotlib.pyplot as plt #standard alias for the industry
```

```python
import numpy as np
import math
```

```python
x = np.linspace(0, 5, 100)  # Adjust the range for each equation

# Define the equations
y1 = np.sqrt(x)
y2 = x ** 3
y3 = 2 ** x
y4 = 1 / (x + 1)

# Create subplots for each equation
plt.figure(figsize=(12, 8))

# Plot the first equation
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('y = âˆšx')
plt.xlabel('x')
plt.ylabel('y')

# Plot the second equation
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('y = x^3')
plt.xlabel('x')
plt.ylabel('y')

# Plot the third equation
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('y = 2^x')
plt.xlabel('x')
plt.ylabel('y')

# Plot the fourth equation
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('y = 1 / (x+1)')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
# Create a single figure with 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot the first equation in the first subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('y = âˆšx')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

# Plot the second equation in the second subplot
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('y = x^3')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

# Plot the third equation in the third subplot
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('y = 2^x')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

# Plot the fourth equation in the fourth subplot
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('y = 1 / (x+1)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```
